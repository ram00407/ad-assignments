```python
m = []
for i in num:
  if str(i).startswith('8'):
    m.append(str(i).startswith('5'))
  else:
    print(i)
```

```python
n=[]
for i in num:
  if str(i).endswith('3'):
    n.append(str(i).endswith('3'))
  else:
    print(i)
```

```python
n=[]
for i in num:
  j = str(i)
  if j[0]=='8':
    a = int('5'+j[1:])
    n.append(a)
  else:
    n.append(i)
```

```python
def separater(x):
  j=[]
  k=[]
  for i in x:
    if i%2==0:
      j.append(i)
    else:
      k.append(i)
  return {'even':j,'odd':k}
```

```python
a=separater([34,43,2,343,647,44,446,23])
```
